# **1. Numerical:**

## **a. Discrete Data:**

It is data that only takes certain values. **Commonly in the form of whole numbers or integers**,

> this is data that can be counted and has a finite number of values.

Imagine you're counting toys. You can count 1 toy, 2 toys, 3 toys, and so on. You can't have half a toy, and there's always a gap between the numbers.

> Discrete data is about counting individual things with gaps between values.

## **b. Continous Data:**

It is data that can be measured. **This data has values that are not fixed and have an infinite number of possible values.**

> These measurements can also be broken down into smaller individual parts.

Think about measuring water in a glass. You can have a little water, a bit more, even a tiny bit more, and it keeps going without any gaps. You can have exactly how much you want, like pouring water from a tap.

> It's not limited to whole numbers; it can be any value within a range.

# **2. Categorical Data:**

Qualitative data that that has no mathematical meaning and  **where the categories have no order or ranking.**

* Categorical data is about grouping things into categories.
* You can't order or measure categorical data like numbers.

> **For example:** Types of Fruits, Marital Status, Genres of Movies, Education Levels etc.

# **3. Ordinal Data:**

It is categorical data that has mathematical meaning and  **where the categories have a clear order or ranking**.

> It is mixture of numerical and categorical data.

For example, movie rating from 1-5. Here, 1-5 are showing categories but have implied relationship. These numbers have mathematical relationship. like in maths, we know 2 > 1. It means if movie got 2 rating, it means movie was good. as compared to if movie got 1 rating it means movie was bad.

## **Okay, variance told us that there is some outlier. How to figure out that outlier????**

Probability mass function is of two types.

1. Binomial Probab mass function (which we have studied above)

2. Poisson (Poa-Son) Probab mass function:

It works on the fact that if we have average value/information of something on a timeperiod, then to estimate the prob of that something on future day.

Definition: The Poisson probability mass function (PMF) tells us the chance or **probability** of seeing a **certain number of rare and random events**, like phone calls or accidents, **in a fixed time** or space, **based on an average rate.**

```python
# 1. Binomial Prob Mass function

from scipy.stats import binom
import matplotlib.pyplot as plt

# Define the number of trials (n) and the probability of success (p).
n, p = 10, 0.5

x = np.arange(0, 10, 0.001)
plt.plot(x, binom.pmf(x, n, p));
```

So, when you see the line "n, p = 10, 0.5" in the code, it's defining the number of trials as 10 (meaning you're repeating the experiment 10 times), and the probability of success as 0.5 (meaning there's a 50% chance of the outcome being a "success" in each trial).

Poisson Probab mass function example:

Suppose my website has 500 visits per day (on average). What is the chance of getting 550?

```python
# 1. Poisson Prob Mass function

from scipy.stats import poisson
import matplotlib.pyplot as plt
import numpy as np

# Define average rate
mu = 500

# Calculate the PMF at x = 550
x_value = 550
pmf_at_x = poisson.pmf(x_value, mu)

#Plotting

x = np.arange(400, 600, 0.5)
plt.plot(x, poisson.pmf(x, mu))
plt.scatter(x_value, pmf_at_x, color='red', label=f'P(X = {x_value})')
plt.legend()
plt.xlabel('Number of Visits')
plt.ylabel('Probability')
plt.title('Poisson Probability Mass Function')
plt.show()
```

# **Moments:**

* It measure the shape of the probability density function.

* **First moment is Mean**

* **Second moment is Variance**

* **Third moment is 'skew ($\gamma$)'**:

It tells how **lopsided (یک طرفہ)** is your distribution.

A distribution with longer tail on left will be left skewed and with longer tail on right will be right skewed:

image.png



Here, left skewed is negative skew and right skew is positive skew.

* Left-skewed (negatively skewed) distributions have a longer left tail and are characterized by having a majority of their values concentrated on the right side of the distribution. This often results in the mean being less than the median.

* Right-skewed (positively skewed) distributions have a longer right tail and are characterized by having a majority of their values concentrated on the left side of the distribution. This can lead to the mean being greater than the median.

* **Forth moment is kurtosis:**

It tells how sharp is your peak of distribution as compared to normal distibution?

---- Higher peaks have higher kurtosis ------

image.png

Here, red distribution has higher kurtosis.

Kurtosis can be interpreted as follows:

1. **Meso-kurtic (Kurtosis = 0):** If the kurtosis value is close to 0, it indicates that the distribution has similar tail behavior as a normal distribution. In this case, the tails are neither too heavy (outliers) nor too light (lack of outliers).

2. **Lepto-kurtic (Kurtosis > 0):** If the kurtosis value is positive, it indicates that the distribution has heavier tails than a normal distribution. This means that the dataset has more extreme values (outliers) than a normal distribution would. A leptokurtic distribution is more peaked at the center and has fatter tails.

3. **Platy-kurtic (Kurtosis < 0):** If the kurtosis value is negative, it indicates that the distribution has lighter tails than a normal distribution. This means that the dataset has fewer extreme values (outliers) than a normal distribution would. A platykurtic distribution is flatter at the center and has thinner tails.

Let's calculate all moments.

```python
import numpy as np
import matplotlib.pyplot as plt

vals = np.random.normal(0, 0.5, 10000)

plt.hist(vals, 50)
plt.show()
```

```python
# FIRST MOMENT

np.mean(vals)
```

```python
# SECOND MOMENT

np.var(vals)
```

```python
# THIRD MOMENT

import scipy.stats as sp
sp.skew(vals)
```
