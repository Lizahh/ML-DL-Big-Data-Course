## **Linear Regression:**

Definition:

**"Fit a line to a set of observations. Use this line to predict unobserved values."**

* Line: y = mx+b

m = slope = It is correlation between the two variables X and Y TIMES the standard deviation in Y, all divided by the SD of X i.e.

        corr(X,Y) / SD(Y) / SD(X)

b = y-intercept = mean of Y minus the slope times the mean of X





* It maximimizes the likelihood of the observed data ==> maximum likelihood estimation.

* R-squared (also known as the coefficient of determination) is commonly used to measure the goodness of fit for a linear regression model.

* It indicates the proportion of the variance in the dependent variable that is explained by the model(independent variables) in the regression model.

The R-squared value ranges from 0 to 1, where:

  * 0 indicates that the model explains none of the variance in the dependent variable.
  * 1 indicates that the model explains all of the variance in the dependent variable.

In the context of linear regression, a higher R-squared value indicates that the independent variables in the model are better at explaining the variation in the dependent variable.

# **-------- Example ----------**

```python
import numpy as np
import seaborn as sns

pageSpeed = np.random.normal(3.0, 1.0, 1000)
purchaseAmount = 100 - (pageSpeed + np.random.normal(0, 0.1, 1000)) * 3

sns.scatterplot(x = pageSpeed, y = purchaseAmount);
```

```python
from scipy import stats

slope, intercept, r_value, p_value, std_err = stats.linregress(pageSpeed, purchaseAmount)
```

```python
# check the r squared value to check the error

r_value ** 2
```

It shows good fit.

```python
# lets plot the line

import matplotlib.pyplot as plt

def predict(X):
  return slope * X + intercept

fitline = predict(pageSpeed)
fitline
```

```python
sns.scatterplot(x = pageSpeed, y = purchaseAmount)
plt.plot(pageSpeed, fitline,c = 'r');
```

# **Polynomial Regression:**

* y = mx+b is first order polynomial. => Linear

* y = ax^2 + bx + c is second order polynomial.

* y = ax^3 + bx^2 + cx + d is third order polynomial.

* Higher order polynomials produce more complex curve which means complexity of model increases by increasing polynomials.

* It can lead to overfitting as you know. So dont use more degree than you need.

* Both OLS and gradient descent can be used when we have multiple features.

1. OLS is a closed-form solution that provides a direct mathematical formula to calculate the optimal coefficients that minimize the sum of squared residuals (the difference between predicted and actual values).

  Gradient descent, on the other hand, is an iterative optimization algorithm. It starts with initial parameter values and iteratively updates them in the direction of the steepest decrease in the cost function (e.g., mean squared error) until convergence is reached.


  2. OLS is primarily used for linear regression, but it can also be applied to other linear models, such as multiple linear regression, polynomial regression, and multivariate regression
  
  Gradient descent is highly versatile and can be used not only in linear regression but also in various machine learning models, including linear regression with one or more features, logistic regression, neural networks, and more.

  3. OLS can be computationally expensive when dealing with a large number of features, as it requires matrix inversion, which can be slow for high-dimensional data.
  
  Gradient descent is generally more scalable and efficient for high-dimensional datasets, as it processes data in smaller batches or one data point at a time, making it suitable for large-scale machine learning tasks.
