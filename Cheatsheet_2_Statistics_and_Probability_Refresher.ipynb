# **2. Piechart:**



* **autopct**='%1.1f%%' means:

  * **'1.1f':** This is a format string specifying how the percentage should be displayed. In this case, it's formatted as a floating-point number with one digit before the decimal point and one digit after the decimal point (e.g., "25.0%").
'%%': The double percentage sign %% is used to display a literal percentage symbol (%) after the number.

*  **startangle:** This parameter specifies the angle at which the first pie slice starts.

  * if you set "startangle" to 0 degrees, you start looking at the pie from the right side. If you set it to 90 degrees, you start looking at the pie from the top. It's like choosing where you want to begin when you look at the pie chart.

```python
import matplotlib.pyplot as plt

# Data
labels = ['Category A', 'Category B', 'Category C', 'Category D']
sizes = [25, 30, 15, 30]  # Percentages

# Create a pie chart
plt.figure(figsize=(6, 6))  # Optional: Set the figure size
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)

# Optional: Add a title
plt.title('Simple Pie Chart')

# Show the chart
plt.show()
```

## **Measuring Covariance:**

---- Covariance is like a measurement that tells us how two things change together compared to their average values---------

* **Think of the two variables as vectors.**

* **Convert these vectors to ---> Vectors of Variance from mean**

(For it,you first need to find how each data point in these lists varies (or differs) from its respective mean (average). This means you calculate the difference between each data point and the average value for that set of data.

For temperatures, if the average is 80 degrees, and one day's temperature is 85 degrees, the variance from the mean for that day would be 85 - 80 = 5 degrees.
Similarly, for ice cream sales, if the average daily sales are 200 units, and one day's sales are 220 units, the variance from the mean for that day would be 220 - 200 = 20 units.)

* **Find Similarity between these two vectors by taking the dot product between them**

* **Divide it by sample size i.e. number of rows in both variables is same ofcourse. So divide by that number**






```python
# FIND CORRELATION BETWEEN PAGE LOADING SPEED AND PURCHASE AMOUNT FROM THAT PAGE

import numpy as np
import seaborn as sns

# STEP 2 ---- FIND VECTORS OF VARIANCE FROM MEAN

def vec_of_var_from_mean(vectr):
  mean_of_vec = np.mean(vectr)
  var_from_mean = [xi - mean_of_vec for xi in vectr]

  return var_from_mean

# STEP 3 ----- TAKE DOT PRODUCT AND DIVIDE BY SAMPLE SIDE

def covariance(X, Y):
  sample_size = len(X)

  cov=  np.dot(vec_of_var_from_mean(X), vec_of_var_from_mean(Y)) / sample_size

  return cov

# STEP 1  --- GIVE THE DATA

pagespeed = np.random.normal(3.0, 1.0, 1000)
purchases = np.random.normal(50.0, 10.0, 1000)/pagespeed

# STEP 4 ---- Call the functions

print("Covariance is: {}".format(covariance(pagespeed, purchases)))

sns.scatterplot(x=pagespeed, y=purchases)
plt.xlabel("Page Loading Speed")
plt.ylabel("Purchase Amount")
plt.title("Scatterplot of Page Loading Speed vs. Purchase Amount");
```

## **Interpreting Covariance:**

* CLOSE TO 0 ----> Not much correlation between both variables

* FAR FROM 0 ----> There is correlation. Doesnt matter +ve or -ve, but there is.

How much far can I go??? Is there any limit???? In our above code, we got -7 covariance.

you want limit? Then convert this covariance to correlation. It ranges from -1 (perfect negative correlation) to 1 (perfect positive correlation), with 0 indicating no linear correlation.

## **Measuring Correlation:**

* **Divide the covariance of both X and Y by the Standard deviation of both X and Y ------------>** GET CORRELATION 

* --- FYI: Converting the data in a range is called normalization ---

Formula:

Correlation Coefficient ( or r) = Covariance(X, Y) / (X * Y)





## **Interpreting Correlation:**



  * Correlation = 1 : Perfect Correlation
  * Correlation = 0 : No correlation at all
  * Correlation = -1: Inverse Correlation

```python
# NORMALIZING COVARIANCE

def correlation(X, Y):
  std_x = X.std()
  std_y = Y.std()

  corr = covariance(X, Y) / std_x / std_y

  return corr

print("Correlation is: {}".format(correlation(pagespeed, purchases)))
```
